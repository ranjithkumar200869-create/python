```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/Titanic_Dataset[1].csv")
print("dataset loaded sucessfully")

```

```python
print("First 5 Rows:")
print(df.head())

print("\nDataset Information:")
df.info()

print("\nStatistical Summary:")
print(df.describe())
```

```python
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())
df['age'].fillna(df['age'].mean(), inplace=True)
if 'cabin' in df.columns.str.lower():
    df.drop('cabin', axis=1, inplace=True)
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
age_series = df['age']
fare_series = df['fare']
survival_series = df['survived']
print("\nage Series Sample:")
print(age_series.head())
```

```python
print("\nPassenger using loc (index 5):")
print(df.loc[5])

print("\nPassenger using iloc (index 10):")
print(df.iloc[10])

print("\nSelected Columns (age, fare, survived):")
print(df[['age', 'fare', 'survived']].head())
```

```python
df.rename(columns={'Sex': 'Gender'}, inplace=True)
print("\nColumn 'Sex' renamed to 'Gender'")
```

```python
print("\nPassengers with age > 30:")
print(df[df['age'] > 30].head())

print("\nPassengers with fare > 50:")
print(df[df['fare'] > 50].head())

print("\nFemale Passengers Who survived:")
print(df[(df['sex'] == 'Female') & (df['survived'] == 1)].head())
```

```python
print("\nTop 5 Highest fare:")
print(df.sort_values(by='fare', ascending=False).head())

print("\nYoungest Passengers:")
print(df.sort_values(by='age').head())
```

```python
print("\nSurvival Rate by sex:")
print(df.groupby('sex')['survived'].mean())

print("\nAverage age by Passenger Class:")
print(df.groupby('pclass')['age'].mean())

print("\nSurvival Count by embarked:")
print(df.groupby('embarked')['survived'].sum())
```

```python
class_info = pd.DataFrame({
    'pclass': [1, 2, 3],
    'Class_Category': ['Upper', 'Middle', 'Lower']})

merged_df = pd.merge(df, class_info, on='pclass')

print("\nMerged Data Sample:")
print(merged_df.head())
```

```python
print("\nAverage age (survived vs Not survived):")
print(df.groupby('survived')['age'].mean())

print("\nHighest fare Paid:", df['fare'].max())
print("Lowest fare Paid:", df['fare'].min())

survival_percentage = df['survived'].mean() * 100
print("\nOverall Survival Percentage: {:.2f}%".format(survival_percentage))

print("\nProject Completed Successfully!")
```
